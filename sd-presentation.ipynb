2. Clean NaN

```
df_NaN = round((df.isnull().sum()/len(df))*100).astype("int")
df_NaN
```

```
# clean data
def process_data(data, nan_rate):
  length = len(data.index)
  dirty_arr = []
  for col in data.columns:
    if data[col].isnull().sum()> nan_rate * length:
      dirty_arr.append(col)
  return data.drop(columns=dirty_arr)
  

df_clean = process_data(df, 0.5)
# df_clean.describe()
df_clean.info()
df_clean.shape
```

```
def find_high_corr(data, corr_min=0.19):
  high_corr = {}
  corr_matrix = data.corr()
  corr_matrix_m = data[data["gender"] == 1].corr()
  corr_matrix_f = data[data["gender"] == 0].corr()
  for c in corr_matrix.columns:
    if (np.abs(corr_matrix.loc['match', c]) > corr_min):
      high_corr[c] = [corr_matrix.loc['match', c], corr_matrix_m.loc['match', c], corr_matrix_f.loc['match', c]]
  high_corr.pop("match")
  df = pd.DataFrame(high_corr, index=["mixed", "male", "female"])  
  return df

#corr_2 = df.corr()
#corr_with_match = corr_2[np.abs(corr_2['match']) > 0.2]['match']
# corr_with_match = corr_2[corr_2['match'] > 0.2]['match']
#corr_with_match
df_high_corr = find_high_corr(df, corr_min=0.2)
```

```
df_f = df[df.gender == 0]
df_h = df[df.gender == 1]

def ratio_positive_decision(data=df):
  numerateur = data.groupby("iid").sum()["dec"]
  denominateur = data.groupby("iid").count()["dec"]
  return np.round((numerateur / denominateur).mean(), 2)


a = ratio_positive_decision()
b = ratio_positive_decision(df_h)
c = ratio_positive_decision(df_f)
data = {'Gender':['Mixte', 'Male', 'Female'],
        'Ratio':[a, b, c]}
sns.catplot(x='Gender', y='Ratio',data=pd.DataFrame(data), kind='bar')


# dec_dataset = new_clean_dataset.groupby("iid").sum()["dec"]/new_clean_dataset.groupby("iid").count()["dec"]
```
